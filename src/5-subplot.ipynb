### Exercise 5: Subplot Layout

```python
# TASK: Create a 2x2 subplot layout.
# Plot a line chart in the first subplot, a bar chart in the second, a scatter plot in the third, and a pie chart in the fourth.
```

```python
import matplotlib.pyplot as plt
import numpy as np

# Data for the charts
x = np.linspace(0, 10, 100)
y = x**2
categories = ['A', 'B', 'C', 'D']
values = [5, 7, 3, 9]
scatter_x = [1, 2, 3, 4, 5]
scatter_y = [2, 4, 6, 8, 10]
labels = ['Python', 'Java', 'C++', 'JavaScript']
sizes = [40, 25, 20, 15]

# Create a 2x2 subplot layout
fig, axs = plt.subplots(2, 2, figsize=(10, 8))

# Plot 1: Line chart (top-left)
axs[0, 0].plot(x, y)
axs[0, 0].set_title('Line Chart: y = x^2')

# Plot 2: Bar chart (top-right)
axs[0, 1].bar(categories, values, color=['red', 'blue', 'green', 'purple'])
axs[0, 1].set_title('Bar Chart')

# Plot 3: Scatter plot (bottom-left)
axs[1, 0].scatter(scatter_x, scatter_y)
for i in range(len(scatter_x)):
    axs[1, 0].annotate(f'({scatter_x[i]}, {scatter_y[i]})', (scatter_x[i], scatter_y[i]), textcoords="offset points", xytext=(0, 5), ha='center')
axs[1, 0].set_title('Scatter Plot with Annotations')

# Plot 4: Pie chart (bottom-right)
axs[1, 1].pie(sizes, labels=labels, autopct='%1.1f%%', startangle=90)
axs[1, 1].set_title('Pie Chart')

# Adjust layout for better spacing
plt.tight_layout()

# Display the plot
plt.show()
```
